## Breve Histórico

A COVID-19 é uma doença infectocontagiosa cujos primeiros casos foram reportados no final de 2019 em Wuhan, na China.

Com sua rápida disseminação ao redor do mundo, em 11 de março de 2020 a COVID-19 foi classificada pela a Organização Mundial da Saúde como uma pandemia.

Ao longo de quase três anos de pandemia, muito conhecimento foi rapidamente construído sobre a doença, com o envolvimento de cientistas e profissionais da saúde de todo o mundo, porém, muito ainda temos a investigar e compreender.

## Causa

A COVID-19 é causada pelo vírus SARS-CoV-2, um tipo de coronavírus humano.

## Transmissão

A transmissão da doença se dá, principalmente pelo contato direto com gotículas (saliva, espirro) de uma pessoa infectada.

## Principais Sintomas

Grande parte dos pacientes são assintomáticos ou apresentam sintomas leves, como dor de garganta, coriza, dor no corpo, perda de olfato e paladar e a maioria evolui clinicamente bem, sem necessidade de atendimento hospitalar.

Em uma parte dos casos há febre, falta de ar com queda da saturação (nível de oxigênio no sangue) que em casos mais severos necessitam de internação com uso de respiradores artificiais.

## Diagnóstico

O padrão-ouro para o diagnóstico da COVID-19 é o exame de PCR realizado geralmente em uma amostra de swab de nasofaringe que detecta a presença do vírus por seu material genético.

Há também testes que detectam o antígeno do vírus e outros que buscam anticorpos específicos IgM e IgG produzidos pelo nosso organismo para combater a infecção.

## Tratamento

Em geral, são tratados os sintomas da doença, na maioria das vezes com uso de analgésicos, antitérmicos, antiinflamatórios, antimicrobianos, com possibilidade de em casos mais severos, ser necessária internação e uso de suprimento de oxigênio e de respiradores artificiais.


## Prevenção


No início da pandemia, a prevenção se deu principalmetne pelo distanciamento social, utilização de máscaras e higiene das mãos com água e sabão ou álcool em gel.

Iniciou-se uma corrida pela vacina e diversas indústrias farmacêuticas desenvolveram imunizantes em tempo *record*. No final de 2020 as primeiras vacinas já estavam sendo aprovadas ao redor do mundo e no início de 2021 as primeiras doses foram aplicadas no Brasil.

### Dicionário de Variáveis

* iso_code = código de 3 letras que identifica um país
* continent = continente no qual o país está localizado geograficamente
* location = nome do país ou localização geográfica considerada
* date = data da observação
* total_cases = total de casos confirmados de COVID-19
* new_cases = novos casos confirmados de COVID-19 
* new_cases_smoothed = média do número de novos casos confirmados de COVID-19 nos últimos 7 dias
* total_deaths = total de mortes causadas pela COVID-19
* new_deaths = novas mortes causadas pela COVID-19
* new_deaths_smoothed = média do número de novas mortes causadas pela COVI-19 nos últimos 7 dias
* total_cases_per_million = total de casos confirmados por 1 milhão de habitantes
* new_cases_per_million = novos casos confirmados por 1 milhão de habitantes
* new_cases_smoothed_per_million = média do número de novos caos confirmados de COVID-19 por 1 milhão de habitantes nos últimos 7 dias
* total_deaths_per_million = total de mortes causadas pela COVID-19, por 1 milhão de habitantes
* new_deaths_per_million = novas mortes causadas pela COVID-19, por 1 milhão de habitantes
* new_deaths_smoothed_per_million = média do número de novas mortes causadas pela COVI-19 nos últimos 7 dias
* reproduction_rate = estimativa em tempo real da taxa de reprodução (R) da COVID-19
* icu_patients = número de pacientes em unidades de terapia intensiva em um dado dia
* icu_patients_per_million = número de pacientes em unidades de terapia intensiva em um dado dia, por 1 milhão de habitantes
* hosp_patients = número de pacientes hospitalizados em um dado dia
* hosp_patients_per_million = número de pacientes hospitalizados em um dado dia, por 1 milhão de habitantes
* weekly_icu_admissions = número de novos pacientes que deram entrada em unidades de terapia intensiva em uma dada semana (dado atual e dos 6 dias anteriores)
* weekly_icu_admissions_per_million = número de novos pacientes que deram entrada em unidades de terapia intensiva em uma dada semana (dado atual e dos 6 dias anteriores), por 1 milhão de habitantes
* weekly_hosp_admissions = número de novos pacientes hospitalizados em uma dada semana (dado atual e dos 6 dias anteriores)
* weekly_hosp_admissions_per_million = número de novos pacientes hospitalizados em uma dada semana (dado atual e dos 6 dias anteriores), por 1 milhão de habitantes
* total_tests = total de testes realizados para COVID-19
* new_tests = novos testes realizados para COVID-19
* total_tests_per_thousand = total de testes realizados para COVID-19, por 1000 habitantes
* new_tests_per_thousand = novos testes realizados para COVID-19, por 1000 habitantes
* new_tests_smoothed = média do número de novos testes realizados para COVID-19 nos últimos 7 dias
* new_tests_smoothed_per_thousand = média do número de novos testes realizados para COVID-19 nos últimos 7 dias por mil habitantes
* positive_rate = taxa de positividade dos testes realizados dada a média móvel dos últimos 7 dias
* tests_per_case = testes realizados por novo caso confirmado de COVID-19 dada a média móvel dos últimos 7 dias
* tests_units = unidades usadas pelo local para relatar seus dados de teste
* total_vaccinations = número total de doses de vacina contra COVID-19 administradas
* people_vaccinated = número total de pessoas que receberam pelo menos uma dose da vacina
* people_fully_vaccinated = número de pessoas que receberam todas as doses de vacina prescritas no protocolo de vacinação utilizado
* total_boosters = número total de doses de reforço da vacinação contra COVID-19 administradas (doses administradas além do número prescrito pelo protocolo de vacinação)
* new_vaccinations = novas doses de vacina contra COVID-19 administradas
* new_vaccinations_smoothed = média do número de novas doses de vacina contra COVID-19 administradas nos últimos 7 dias
* total_vaccinations_per_hundred = número total de doses de vacina contra COVID-19 administradas, por 100 habitantes
* people_vaccinated_per_hundred = número total de pessoas que receberam pelo menos uma dose da vacina, por 100 habitantes
* people_fully_vaccinated_per_hundred = número de pessoas que receberam todas as doses de vacina prescritas no protocolo de vacinação utilizado, por 100 habitantes
* total_boosters_per_hundred = número total de doses de reforço da vacinação contra COVID-19 administradas (doses administradas além do número prescrito pelo protocolo de vacinação), por 100 habitantes
* new_vaccinations_smoothed_per_million = média do número de novas doses de vacina contra COVID-19 administradas nos últimos 7 dias, por 1 milhão de habitantes
* new_people_vaccinated_smoothed = média do número diário de pessoas recebendo a primeira dose da vacina, nos últimos 7 dias
* new_people_vaccinated_smoothed_per_hundred = média do número diário de pessoas recebendo a primeira dose da vacina, nos últimos 7 dias, por 100 habitantes
* stringency_index = índice de Severidade de Resposta do Governo: medida composta com base em 9 indicadores de resposta, incluindo fechamento de escolas, fechamento de locais de trabalho e proibições de viagens, reescalonados para um valor de 0 a 100 (100 = resposta mais rigorosa)
* population = população do país
* population_density = número de pessoas dividido pela área do território do  país
* median_age = média de idade da população projetada para 2020 pelas Nações Unidas
* aged_65_older = população acima de 65 anos de idades
* aged_70_older = população acima de 70 anos de idade
* gdp_per_capita = PIB - Produto interno bruto do país em paridade de poder de compra
* extreme_poverty = população vivendo em extrema pobreza
* cardiovasc_death_rate = taxa de morte por doença cardiovascular em 2017
* diabetes_prevalence = prevalência de diabetes (porcentagem na população entre 20 e 79 anos) em 2017
* female_smokers = número de mulheres fumantes
* male_smokers = número de homens fumantes
* handwashing_facilities = população com acesso a saneamento básico
* hospital_beds_per_thousand = número de leitos hospitalares por 1000 habitantes
* life_expectancy = expectativa de vida ao nascer, em 2019
* human_development_index = IDH - Índice de Desenvolvimento Humano que mede o desempenho médio em três dimensões básicas do desenvolvimento humano – uma vida longa e saudável, conhecimento e um padrão de vida decente
* excess_mortality = diferença percentual entre o número relatado de mortes semanais ou mensais em 2020–2021 e o número projetado de mortes para o mesmo período com base nos anos anteriores
* excess_mortality_cumulative = diferença percentual entre o número acumulado de mortes desde 1º de janeiro de 2020 e as mortes projetadas cumulativas para o mesmo período com base nos anos anteriores
* excess_mortality_cumulative_absolute = diferença cumulativa entre o número relatado de mortes desde 1º de janeiro de 2020 e o número projetado de mortes para o mesmo período com base nos anos anteriores
* excess_mortality_cumulative_per_million = Diferença cumulativa entre o número relatado de mortes desde 1º de janeiro de 2020 e o número projetado de mortes para o mesmo período com base nos anos anteriores, por 1 milhão de habitantes

```
# Selecionar os dados de PIB per capita , IDH , Pobreza Extrema e Expectativa de Vida dos países e o total de mortes e de casos na data mais atual

''
df_socecon = df.loc[df.date == Data_mais_atual, ["location", "total_cases", "total_deaths", "gdp_per_capita", "human_development_index", "extreme_poverty", "life_expectancy"]].loc[df.continent != "NA"]
df_socecon.head()
```

Verificamos aqui que cerca de metade dos dados estão ausentes para a variável extrema pobreza, então não iremos utilizá-la para as demais análises

```
# Plotar um gráfico de dispersão considerando o PIB dos países (gdp_per_capita)) pelo número total de mortos, para a data mais recente

fig, ax = plt.subplots(figsize = (8, 5))

df_socecon.plot(x = "gdp_per_capita", y = "total_deaths", kind = "scatter", color = "orange", ax=ax)

ax.set_title("Dispersão do Total de Mortes considerando o PIB per capita de cada país", fontsize = 15)

ax.set_ylabel("Milhões de Pessoas", fontsize = 15)
ax.set_xlabel("PIB per capita", fontsize = 15)
plt.tight_layout()
```

Podemos verificar visualmente que alguns valores são bem discrepantes e podem ser *outliers*, mas vamos testar uma abordagem criando um dado de taxa de mortalidade (total mortes/total casos), ao invés de tratar os outliers

```
df_socecon['taxa_morte'] = None
```

```
df_socecon['taxa_morte'] = 100 * (df_socecon.total_deaths/df_socecon.total_cases)
df_socecon.head()
```

```
# Plotar um novo gráfico de dispersão considerando o PIB dos países (gdp_per_capita)) agora em função da taxa de morte que calculamos, para a data mais recente

fig, ax = plt.subplots(figsize = (8, 5))

df_socecon.plot(x = "gdp_per_capita", y = "taxa_morte", kind = "scatter", color = "purple", ax=ax)

ax.set_title("Dispersão da Taxa de Mortes considerando o PIB per capita de cada país", fontsize = 15)

ax.set_ylabel("Taxa de Mortes", fontsize = 15)
ax.set_xlabel("PIB per capita", fontsize = 15)
plt.tight_layout()
```

```
# Vamos remover os dados em que a taxa de morte for maior que 5%, buscando retirar alguns outliers (apenas visualmente, sem utilizar métodos estatísticos)

df_clean = df_socecon.copy()
df_clean.drop(df_clean[df_clean.taxa_morte > 5].index, axis = 0, inplace = True)
```

```
# Verificando o novo tamanho do dataset limpo
df_clean.shape
```

```
# Plotar um novo gráfico de dispersão considerando o novo dataset o PIB dos países (gdp_per_capita)) agora em função da taxa de morte que calculamos, para a data mais recente

fig, ax = plt.subplots(figsize = (8, 5))

df_clean.plot(x = "gdp_per_capita", y = "taxa_morte", kind = "scatter", color = "purple", ax=ax)

ax.set_title("Dispersão do Taxa de Mortes considerando o PIB per capita de cada país", fontsize = 15)

ax.set_ylabel("Taxa de Mortes", fontsize = 15)
ax.set_xlabel("PIB per capita", fontsize = 15)
plt.tight_layout()
```

```
# Plotar um gráfico de dispersão considerando o Índice de Desenvolvimento Humano - IDH dos países pela taxa de mortos, para a data mais recente

fig, ax = plt.subplots(figsize = (8, 5))

df_clean.plot(x = "human_development_index", y = "taxa_morte", kind = "scatter", color = "purple", ax=ax)

ax.set_title("Dispersão do Taxa de Mortes considerando o IDH de cada país", fontsize = 15)

ax.set_ylabel("Taxa de Mortes", fontsize = 15)
ax.set_xlabel("IDH", fontsize = 15)
plt.tight_layout()
```

```
# Plotar um gráfico de dispersão considerando a Expectativa de vida dos países pela taxa de mortos, para a data mais recente

fig, ax = plt.subplots(figsize = (8, 5))

df_clean.plot(x = "life_expectancy", y = "taxa_morte", kind = "scatter", color = "purple", ax=ax)

ax.set_title("Dispersão do Taxa de Mortes considerando a Expectativa de Vida de cada país", fontsize = 15)

ax.set_ylabel("Taxa de Mortes", fontsize = 15)
ax.set_xlabel("Expectativa de Vida", fontsize = 15)
plt.tight_layout()
```

Como dissemos acima, não vamos gerar o gráfico da extrema pobreza devida a alta quantidade de valores ausentes.

Podemos inferir, a partir dos gráficos gerados, que não há uma relação aparente entre a riqueza econômica e qualidade de vida nos países e a porcentagem de mortes em relação ao total de casos, o que demonstra como a doença pôde ser implacável até nos locais onde há mais recursos financeiros e tecnológicos, como Europa e EUA.
